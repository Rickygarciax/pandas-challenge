```python
# Dependencies and Setup
import pandas as pd 
```

```python
# File to Load 
highschools_df = "../Resources/schools_complete.csv"
students_df = "../Resources/students_complete.csv"

# Read CSV Files 
highschools_df = pd.read_csv(highschools_df)
students_df = pd.read_csv(students_df)
```

```python
# Preview Highschools Data
highschools_df.head(100000) 
```

```python
# Preview Students Data
students_df.head() 
```

```python
# Combine the Data into one data frame
highschool_students_df = pd.merge(students_df, highschools_df, how="left", on=["school_name", "school_name"])
highschool_students_df.head() 
```

```python
# Calculate the total number of unique schools
diff_schools = len(highschool_students_df["School ID"].unique())
diff_schools 
```

```python
# Calculate the total number of students 
num_students = highschool_students_df["student_name"].count()
num_students
```

```python
# Calculate the total budget
total_budget = highschool_students_df["budget"].sum()
total_budget
```

```python
# Calculate the avg Math score
avg_math_score = highschool_students_df["math_score"].mean()
round(avg_math_score)
```

```python
# Calculate the avg Reading score
avg_reading_score = highschool_students_df["reading_score"].mean()
round(avg_reading_score)
```

```python
# Calculate the percentage of students who passed math 
passing_math_count = highschool_students_df[(highschool_students_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(num_students) * 100
round(passing_math_percentage)
```

```python
# Calculate the percentage of students who passed reading
passing_reading_count = highschool_students_df[(highschool_students_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(num_students) * 100
round(passing_reading_percentage)
```

```python
# Calculate the percentage of students that passed math and reading
passing_math_reading_count = highschool_students_df[
    (highschool_students_df["math_score"] >= 70) & (highschool_students_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(num_students) * 100
round(overall_passing_rate)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame([
    {"Total Schools": diff_schools,
    "Total Students": num_students,
    "Total Budget": total_budget,
    "Average Math Schore": avg_math_score,
    "Average Reading Schore": avg_reading_score,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing": overall_passing_rate}])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
schools_grouped =highschool_students_df["school_name"]
school_names = schools_grouped.unique()
school_names
```

```python
# Select school type
school_types = highschools_df.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the number of students per school
per_school_counts = highschool_students_df.groupby(["school_name"]).count()["student_name"] 
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = highschool_students_df.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
```

```python
# Calculate the average test scores
per_school_math = highschool_students_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = highschool_students_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = highschool_students_df[(highschool_students_df["math_score"] >= 70)]
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading =  highschool_students_df[(highschool_students_df["reading_score"] >= 70)]
school_passing_reading
```

```python
# Calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = highschool_students_df[
    (highschool_students_df["reading_score"] >= 70) & (highschool_students_df["math_score"] >= 70)
]
passing_math_and_reading
```

```python
# Calculate the passing rates
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

school_summary_unformatted = pd.DataFrame({"School Names" : school_names,
"School Type" : school_types,
"School Count" : per_school_counts,
"Total School Budget" : per_school_budget,
"Per Student Budget" : per_school_capita,
"Average Math Score" : per_school_math,
"Average Reading Score" : per_school_reading, 
"% Passing Math" : per_school_passing_math,
"% Passing Reading" : per_school_passing_reading,
"% Overall Passing" : overall_passing_rate})

per_school_summary = school_summary_unformatted.copy()

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary.reset_index(drop = True, inplace = True)
per_school_summary
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["$585", "$585-630", "$630-645", "$645-680"]
labels
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary_unformatted.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins = spending_bins, labels = labels)
school_spending_df.head() 

```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Scores" : spending_math_scores,
    "Average Reading Score" : spending_reading_scores,
    "% Passing Math" : spending_passing_math,
    "% Passing Reading" : spending_passing_reading,
    "% Overall Passing" : overall_passing_spending})

# Display results
spending_summary
```
